### KMeans Clustering

```python
#numerical_data
filtereddata= loan_data.select_dtypes(include= np.number)
filtereddata.head(3)
```

```python
wcss = [] #this stores the within clusters sum of squares of a particular clustering

for i in tqdm(range(1,11)):
    model= KMeans(n_clusters=i, init='k-means++',  random_state=0)
    model.fit(filtereddata)
    inertia=model.inertia_
    wcss.append(inertia)
```

##### Elobow Method
- use elbow method to select the best inertia

```python
sns.lineplot(x=list(range(1,11)), y=wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

- The cluster with the smallest inertia will be used. 
- From the plot  4 has the clearest lowest inertia so we will use 4 clusters

```python
model= KMeans(n_clusters=4, init='k-means++', random_state=0)
model.fit(filtereddata)
filtereddata['label']=model.labels_

#plot the clusters

plt.figure(figsize=[10,8])
sns.scatterplot(data= filtereddata, x='MonthlyIncome', y='Amountof_Loan', hue='label')
plt.title('Clustering results with 4 clusters')
plt.show()
```

```python
plt.figure(figsize=[10,8])
sns.scatterplot(data= filtereddata, x='OnTimeProsperPayments', y='Amountof_Loan', hue='label')
plt.title('Clustering results with 4 clusters')
plt.show()
```

#### Hierachical Clustering 

```python
model= AgglomerativeClustering(n_clusters=4, linkage='average')
model.fit(filtereddata)
filtereddata['labelH']=model.labels_

#plot the clusters

plt.figure(figsize=[10,8])
sns.scatterplot(data= filtereddata, x='MonthlyIncome', y='Amountof_Loan', hue='labelH')
plt.title('Clustering results with 4 clusters')
plt.show()
```

```python
plt.figure(figsize=[10,8])
sns.scatterplot(data= filtereddata, x='OnTimeProsperPayments', y='Amountof_Loan', hue='labelH')
plt.title('Clustering results with 4 clusters')
plt.show()
```

#### DBSCAN 

```python
model= DBSCAN( eps=11, min_samples=6)
model.fit(filtereddata)
filtereddata['labelC']=model.labels_

#plot the clusters

plt.figure(figsize=[10,8])
sns.scatterplot(data= filtereddata, x='MonthlyIncome', y='Amountof_Loan', hue='labelC')
plt.title('Clustering results with DBSCAN')
plt.show()
```

```python
plt.figure(figsize=[10,8])
sns.scatterplot(data= filtereddata, x='OnTimeProsperPayments', y='Amountof_Loan', hue='labelC')
plt.title('Clustering results with 4 DBSCAN')
plt.show()
```
